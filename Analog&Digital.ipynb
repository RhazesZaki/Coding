```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Parameter Sinyal
frekuensi = 5        # Frekuensi sinyal (5 Hz)
# frekuensi adalah suatu gelombang menunjukkan seberapa cepat gelombang tersebut berosilasi dalam satu detik. Dalam hal ini, 5 Hz berarti gelombang tersebut berosilasi sebanyak 5 kali dalam satu detik.
amplitudo = 1        # Amplitudo sinyal
#amplitudo adalah nilai maksimum dari gelombang. Dalam hal ini, amplitudo 1 berarti gelombang mencapai puncaknya pada nilai 1 dan lembahnya pada nilai -1.
fs_analog = 1000     # Sampling tinggi untuk simulasi "Analog"
#fs_analog adalah frekuensi sampling yang sangat tinggi, yang digunakan untuk mensimulasikan sinyal analog yang kontinu. Dengan fs_analog sebesar 1000 Hz, kita mendapatkan representasi yang sangat halus dari sinyal asli.
fs_digital = 20      # Sampling rendah untuk representasi "Digital" (Diskrit)
#fs_digital adalah frekuensi sampling yang lebih rendah, yang digunakan untuk mensimulasikan sinyal digital yang diskrit. Dengan fs_digital sebesar 20 Hz, kita hanya mendapatkan 20 sampel per detik, yang dapat menyebabkan distorsi atau kehilangan informasi jika frekuensi sinyal asli lebih tinggi dari 10 Hz (sesuai dengan Teorema Nyquist).
durasi = 1           # Durasi dalam detik
#durasi adalah panjang waktu selama sinyal akan disimulasikan. Dalam hal ini, durasi 1 detik berarti kita akan melihat sinyal selama satu detik penuh, yang cukup untuk melihat beberapa siklus dari sinyal 5 Hz (karena 5 siklus per detik).

# 2. Membuat Sumbu Waktu (Time Axis)
t_analog = np.linspace(0, durasi, fs_analog)
t_digital = np.linspace(0, durasi, fs_digital)

# 3. Membuat Sinyal Sinusoidal
# Rumus: y = A * sin(2 * pi * f * t)
y_analog = amplitudo * np.sin(2 * np.pi * frekuensi * t_analog)
y_digital = amplitudo * np.sin(2 * np.pi * frekuensi * t_digital)

# 4. Visualisasi
plt.figure(figsize=(12, 6))

# Plot Sinyal Analog
plt.subplot(2, 1, 1)
plt.plot(t_analog, y_analog, color='blue', label='Sinyal Analog (Kontinu)')
plt.title('Simulasi Sinyal Analog vs Digital')
plt.ylabel('Amplitudo')
plt.legend()
plt.grid(True)

# Plot Sinyal Digital (Diskrit)
plt.subplot(2, 1, 2)
plt.stem(t_digital, y_digital, linefmt='red', markerfmt='ro', basefmt=" ", label='Sinyal Digital (Diskrit)')
plt.plot(t_analog, y_analog, color='blue', alpha=0.2, linestyle='--') # Garis bantu analog
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data sampel sederhana (misal: 0, 1, 1, 0, 1)
data = [0, 1, 1, 0, 1, 0]
t = np.arange(len(data))

plt.figure(figsize=(10, 5))

# Versi 1: Garis tegak (Diskrit/Sampling)
plt.subplot(1, 2, 1)
plt.vlines(t, 0, data, colors='blue', linestyles='solid') # Garis tegak
plt.plot(t, data, 'bo') # Titik di ujungnya
plt.title("Bentuk Garis Tegak\n(Sinyal Diskrit)")
plt.ylim(-0.2, 1.2)
plt.grid(True, alpha=0.3)

# Versi 2: Garis patah-patah (Digital/Square)
plt.subplot(1, 2, 2)
plt.step(t, data, where='post', color='red', linewidth=2) # Garis tangga
plt.title("Bentuk Garis Kotak\n(Sinyal Digital Biner)")
plt.ylim(-0.2, 1.2)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
